# 1. Check whether there is any relationship between the amount spent for advertising through five different companies and profit generated

```python
df.corr()
```

```python
sns.pairplot(df,x_vars=['Add1(in Thousands)', 'Add2(in Thousands)','Add3(in Thousands)', 'Add4(in Thousands)', 'Add5(in Thousands)'],
y_vars=['Profit(in Thousands)'])
plt.title('Relationship between Advertising expenses and profit',fontsize=20,x=-1,y=1)
```

```python
add1=df['Add1(in Thousands)']
add2=df['Add2(in Thousands)']
add3=df['Add3(in Thousands)']
add4=df['Add4(in Thousands)']
add5=df['Add5(in Thousands)']
profit=df['Profit(in Thousands)']
pearsonr_coefficient,p_value=pearsonr(add1,profit)
print("Pearson Correlation Coefficient for add1: %0.3f"%(pearsonr_coefficient))
pearsonr_coefficient,p_value=pearsonr(add2,profit)
print("Pearson Correlation Coefficient for add2: %0.3f"%(pearsonr_coefficient))
pearsonr_coefficient,p_value=pearsonr(add3,profit)
print("Pearson Correlation Coefficient for add3: %0.3f"%(pearsonr_coefficient))
pearsonr_coefficient,p_value=pearsonr(add4,profit)
print("Pearson Correlation Coefficient for add4: %0.3f"%(pearsonr_coefficient))
pearsonr_coefficient,p_value=pearsonr(add5,profit)
print("Pearson Correlation Coefficient for add5: %0.3f"%(pearsonr_coefficient))
```

# Insights

From the above Correlation Analysis we can say that, all the ads are having strong positive correlation to the profit. Correlation takes values between -1 to +1, 
wherein values close to +1 represents strong positive correlation and values close to -1 represents strong negative correlation

# 2. Build a regression model to predict the profit from the given data

```python
corrmatrix=df.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,vmin=0.98,vmax=1,annot=True,linewidth=0.2)
plt.title('Relationship between Add expenses and profit',fontsize=15,y=1)
```

Feature reduction

```python
y=df['Profit(in Thousands)']
x=df.drop(['Date','Profit(in Thousands)'],axis=1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
```

Checking for accuracy for model using predictions and test values

```python
#1)Mean Square Error
from sklearn.metrics import mean_squared_error
print('MSE is:',mean_squared_error(y_test,predictions))
```

```python
#2)Coefficient of determination
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
print("Coefficient of determination is :",r2)
```

# Insights

 MSE indicating a very good model and the coefficient of determination is 1 which indicates a perfect model
