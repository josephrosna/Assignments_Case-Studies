# Case study on Presentation of data

Do the following operations on ‘cars_data’ dataset:
1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset

```python
cars = 'cars_data.csv'
df = pd.read_csv(cars)
```

```python
df
```

# 2. Check the null values

```python
df.isnull().sum()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(x='Buyer Gender',data=df)
plt.title('Male VS Female Buyers',size=15)
```

# 4. Top 5 cars

```python
df1=df.sort_values(by=['Sale Price'],ascending=False).head(5)
df1
```

```python
df2=df1[['Make','Model','Country','Sale Price']]
df2
```

# 5. Least 5 cars based on their Resell price.

```python
df3=df.sort_values(by=['Resell Price'],ascending=True).head(5)
df3
```

```python
df4=df3[['Make','Model','Sale Price','Resell Price']]
df4
```
