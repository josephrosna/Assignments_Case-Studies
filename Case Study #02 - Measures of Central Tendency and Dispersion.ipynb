# Case Study on Measures of Central Tendency and Dispersion

An Institution wishes to find out their student’s ability in maths, reading and writing skills. The Institution wants to do an exploratory study to check the following information.

1. Find out how many males and females participated in the test.
2. What do you think about the students' parental level of education?
3. Who scores the most on average for math, reading and writing based on
    ● Gender
    ● Test preparation course
4. What do you think about the scoring variation for math, reading and writing based on
    ● Gender
    ● Test preparation course
5. The management needs your help to give bonus points to the top 25% of students based on their maths score, so how will you help the management to achieve this.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset

```python
student_performance='StudentsPerformance.csv'
df=pd.read_csv(student_performance)
df
```

# 1. Find out how many males and females participated in the test

```python
count=df.groupby('gender').size()
print('Count of Male and female participants : ')
count
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(x='gender',data=df)
plt.title('Male VS Female Participants',size=15)
```

From this graph we can see that, there is more than 500 female students and only 482 male students participated in the test. 


# 2. What do you think about the students' parental level of education?

```python
plt.figure(figsize=(15,15))
graph=sns.countplot(x='parental level of education',data=df)
plt.title('Parental level of Education',size=30)
for i in graph.patches:
        graph.annotate('{}'.format(i.get_height()), (i.get_x()+0.30, i.get_height()+2))
```

From the graph, most of the parents were went to some college which is around 226
And only 56 parents were having any master's degree and in total 375 parents are having only high school education

# 3. Who scores the most on average for math, reading and writing  


```python

#gender
df_gen_std=df.groupby('gender',as_index=False)[["math score","reading score","writing score"]].std()

#test preparation course

df_course_std=df.groupby('test preparation course',as_index=False)[["math score","reading score","writing score"]].std()

print('Variation in score values of MALE and FEMALE participants are:\n')
df_gen_std
```

```python
print('Variation in score values as per test preparation course completed Scores and none completed are:\n')
df_course_std
```

# 4. What do you think about the scoring variation for math, reading and writing

```python
fig, ax = plt.subplots(1,2,figsize=(15,10))

ax[0].set_ylim(0, 20)
ax[1].set_ylim(0, 20)
plt.ylabel("Standard Deviation",fontsize=15)

graph=df_gen_std[["gender","math score","reading score","writing score"]].plot(x="gender", kind="bar",ax=ax[0])
graph.set_title('Variation based on gender',fontsize=20,pad=15)

for p in graph.patches:
        graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.0, p.get_height()+.5))
graph=df_course_std[["test preparation course","math score","reading score","writing score"]].plot(x="test preparation course", kind="bar",ax=ax[1])
graph.set_title('Variation based on test preparation course',fontsize=20,pad=15)
for p in graph.patches:
        graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.0, p.get_height()+.5))
```

From variation based on the gender, female students have the highest score than the male students.

And from based on test preparation course male students are the ones who scored more

# 5. The management needs your help to give bonus points to the top 25% of students based on their maths score, so how will you help the management to achieve this

```python
df_math_sort=df.sort_values('math score')
df_math_sort=df_math_sort.tail(250)
pd.set_option('display.max_rows',250)
df_math_sort
```

The top 25% only include 250/1000 students but it is seen that the least score is 77 which is scored by 5 more students as per the data. Hence the management needs to consider 255 students for bonus points as per the data
